```python
# Cell 3: 匯入庫與設定種子 (Seed)
# -----------------------------------------------------------------------
import random
import numpy as np
import pandas as pd

import torch
from torch import nn
from transformers.models.deberta_v2.modeling_deberta_v2 import DebertaV2ForSequenceClassification
from torch.utils.data import Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments,
    DataCollatorWithPadding,
)
import re
import html
from tqdm.auto import tqdm

RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
torch.cuda.manual_seed_all(RANDOM_SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device) # 🌟 確保輸出是 cuda
```

```python
# 檔案路徑
train_path = "/content/train.csv"
test_path = "/content/test.csv"
sample_sub_path = "/content/sample_submission.csv"

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
sample_sub = pd.read_csv(sample_sub_path)

# 補 NaN 為空字串
for col in ["keyword", "location", "text"]:
    train_df[col] = train_df[col].fillna("")
    test_df[col] = test_df[col].fillna("")


def build_full_text(row):
    return f"{row['keyword']} {row['text']}"


def clean_text_advanced(text: str) -> str:
    # 1. 還原 HTML entity
    text = html.unescape(text)
    # 2. 移除網址
    text = re.sub(r"http\S+|www\.\S+", "", text)
    # 3. 移除使用者提及
    text = re.sub(r"@[a-zA-Z0-9_]+", "", text)
    # 4. 處理 Hashtag
    text = re.sub(r"#([a-zA-Z0-9_]+)", r"\1", text)
    # 5. 合併多個空白
    text = re.sub(r"\s+", " ", text).strip()
    return text

# 建立並清理 full_text
train_df["full_text"] = train_df.apply(build_full_text, axis=1).apply(clean_text_advanced)
test_df["full_text"] = test_df.apply(build_full_text, axis=1).apply(clean_text_advanced)


# -----------------------------------------------------------
# 🌟 LLM 特徵提取部分 (Simulation)
# -----------------------------------------------------------

def get_llm_features_simulated(text):
    """
    此為模擬 LLM 呼叫，在實際競賽中需替換為真實 API 呼叫。
    提取兩個特徵: 1. 情感分數 2. 是否涉及暴力
    """
    text_lower = text.lower()

    if "bomb" in text_lower or "attack" in text_lower or "kill" in text_lower:
        sentiment = -0.8
        is_violent = 1.0 # 必須是浮點數
    elif "fire" in text_lower or "flood" in text_lower or "storm" in text_lower:
        sentiment = -0.5
        is_violent = 0.0
    else:
        sentiment = 0.1
        is_violent = 0.0

    # 回傳 float 類型
    return {"sentiment_score": sentiment, "is_violent": is_violent}


# 獲取訓練集和測試集的 LLM 特徵
llm_features_train = []
for text in tqdm(train_df['full_text'], desc="Simulating Train LLM Features"):
    llm_features_train.append(get_llm_features_simulated(text))

llm_features_test = []
for text in tqdm(test_df['full_text'], desc="Simulating Test LLM Features"):
    llm_features_test.append(get_llm_features_simulated(text))

# 將特徵加入 DataFrame
llm_feat_df_train = pd.DataFrame(llm_features_train)
train_df = pd.concat([train_df, llm_feat_df_train], axis=1)

llm_feat_df_test = pd.DataFrame(llm_features_test)
test_df = pd.concat([test_df, llm_feat_df_test], axis=1)


print("清理後的訓練數據範例:")
print(train_df[['text', 'full_text', 'sentiment_score', 'is_violent']].head())
```

```python
X = train_df["full_text"].values
y = train_df["target"].values

# LLM 特徵
X_llm = train_df[['sentiment_score', 'is_violent']].values


# 劃分訓練集和驗證集 (10% 驗證集)
X_train, X_valid, y_train, y_valid, X_llm_train_split, X_llm_valid = train_test_split(
    X,
    y,
    X_llm, # 🌟 將 LLM 特徵也一同分割
    test_size=0.1,
    random_state=RANDOM_SEED,
    stratify=y
)

print("Train size:", len(X_train))
print("Valid size:", len(X_valid))
print(f"LLM Train 特徵維度: {X_llm_train_split.shape}")

MODEL_NAME = "microsoft/deberta-v3-base"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
MAX_LEN = 128

# -----------------------------------------------------------
# 🌟 1. 修改 DisasterDataset 類別以處理 LLM Features
# -----------------------------------------------------------
class DisasterDataset(Dataset):
    def __init__(self, texts, llm_features=None, labels=None, tokenizer=None, max_len=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.llm_features = llm_features

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])

        enc = self.tokenizer(
            text,
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )

        item = {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
        }

        if self.llm_features is not None:
            # 🌟 新增 LLM 特徵並轉為 float Tensor
            item["llm_features"] = torch.tensor(self.llm_features[idx], dtype=torch.float)

        if self.labels is not None:
            item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)

        return item

# -----------------------------------------------------------
# 🌟 2. 建立 CustomDataCollator 以批次化 LLM Features
# -----------------------------------------------------------
class CustomDataCollator(DataCollatorWithPadding):
    def __call__(self, features):
        # 提取 LLM features
        llm_features = [f.pop("llm_features") for f in features]
        llm_features = torch.stack(llm_features) # 轉換成 Tensor 批次

        # 讓父類處理主要的 text features
        batch = super().__call__(features)

        # 將 LLM features 加回批次
        batch["llm_features"] = llm_features
        return batch


# 建立 Dataset 物件 (使用分割後的 LLM 特徵)
train_dataset = DisasterDataset(X_train, X_llm_train_split, y_train, tokenizer, MAX_LEN)
valid_dataset = DisasterDataset(X_valid, X_llm_valid, y_valid, tokenizer, MAX_LEN)

# 測試集 LLM 特徵 (從 Cell 4 獲取)
X_llm_test = test_df[['sentiment_score', 'is_violent']].values
test_dataset = DisasterDataset(test_df["full_text"].values, X_llm_test, labels=None, tokenizer=tokenizer, max_len=MAX_LEN)

# 實例化自定義 Data Collator
custom_data_collator = CustomDataCollator(tokenizer=tokenizer)
```

```python
X = train_df["full_text"].values
y = train_df["target"].values

# LLM 特徵
X_llm = train_df[['sentiment_score', 'is_violent']].values


# 劃分訓練集和驗證集 (10% 驗證集)
X_train, X_valid, y_train, y_valid, X_llm_train_split, X_llm_valid = train_test_split(
    X,
    y,
    X_llm, # 🌟 將 LLM 特徵也一同分割
    test_size=0.1,
    random_state=RANDOM_SEED,
    stratify=y
)

print("Train size:", len(X_train))
print("Valid size:", len(X_valid))
print(f"LLM Train 特徵維度: {X_llm_train_split.shape}")

MODEL_NAME = "microsoft/deberta-v3-base"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
MAX_LEN = 128

# -----------------------------------------------------------
# 🌟 1. 修改 DisasterDataset 類別以處理 LLM Features
# -----------------------------------------------------------
class DisasterDataset(Dataset):
    def __init__(self, texts, llm_features=None, labels=None, tokenizer=None, max_len=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len
        self.llm_features = llm_features

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])

        enc = self.tokenizer(
            text,
            truncation=True,
            max_length=self.max_len,
            return_tensors="pt"
        )

        item = {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
        }

        if self.llm_features is not None:
            # 🌟 新增 LLM 特徵並轉為 float Tensor
            item["llm_features"] = torch.tensor(self.llm_features[idx], dtype=torch.float)

        if self.labels is not None:
            item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)

        return item

# -----------------------------------------------------------
# 🌟 2. 建立 CustomDataCollator 以批次化 LLM Features
# -----------------------------------------------------------
class CustomDataCollator(DataCollatorWithPadding):
    def __call__(self, features):
        # 提取 LLM features
        llm_features = [f.pop("llm_features") for f in features]
        llm_features = torch.stack(llm_features) # 轉換成 Tensor 批次

        # 讓父類處理主要的 text features
        batch = super().__call__(features)

        # 將 LLM features 加回批次
        batch["llm_features"] = llm_features
        return batch


# 建立 Dataset 物件 (使用分割後的 LLM 特徵)
train_dataset = DisasterDataset(X_train, X_llm_train_split, y_train, tokenizer, MAX_LEN)
valid_dataset = DisasterDataset(X_valid, X_llm_valid, y_valid, tokenizer, MAX_LEN)

# 測試集 LLM 特徵 (從 Cell 4 獲取)
X_llm_test = test_df[['sentiment_score', 'is_violent']].values
test_dataset = DisasterDataset(test_df["full_text"].values, X_llm_test, labels=None, tokenizer=tokenizer, max_len=MAX_LEN)

# 實例化自定義 Data Collator
custom_data_collator = CustomDataCollator(tokenizer=tokenizer)
```

```python
# Cell 6: 修正後的模型架構與指標函數

from transformers.models.deberta_v2.modeling_deberta_v2 import DebertaV2ForSequenceClassification
from torch import nn

# -----------------------------------------------------------
# 🌟 3. 自定義 DeBERTa 模型類別，實現特徵融合 (已修正 Dropout)
# -----------------------------------------------------------
class DebertaWithLLMFeatures(DebertaV2ForSequenceClassification):
    def __init__(self, config):
        super().__init__(config)

        # LLM 特徵數量 (2 個：sentiment_score, is_violent)
        self.num_llm_features = 2

        deberta_hidden_size = config.hidden_size # Base 版為 768

        # 新增一個融合層：輸入維度 = DeBERTa 隱藏層大小 + LLM 特徵數
        self.llm_fusion_layer = nn.Sequential(
            nn.Linear(deberta_hidden_size + self.num_llm_features, deberta_hidden_size),
            nn.ReLU(),
            # 🌟 修正: 使用 config.hidden_dropout_prob 代替不存在的 config.classifier_dropout
            nn.Dropout(config.hidden_dropout_prob)
        )

    def forward(self, input_ids=None, attention_mask=None, token_type_ids=None,
                llm_features=None, labels=None, **kwargs):

        # 1. 執行 DeBERTa Base Model 的前向傳播
        outputs = self.deberta(
            input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids,
            return_dict=True,
        )

        # 2. 獲取 DeBERTa 的 CLS token 輸出
        sequence_output = outputs.last_hidden_state
        cls_output = sequence_output[:, 0]

        # 3. 融合 LLM 特徵
        if llm_features is not None:
            # 拼接：[CLS Output, LLM Features]
            fused_output = torch.cat((cls_output, llm_features), dim=1)

            # 跑過融合層
            fused_output = self.llm_fusion_layer(fused_output)

            # 4. 使用融合後的輸出進行最終分類 (接上 DeBERTa 的分類頭)
            logits = self.classifier(fused_output)
        else:
            # 如果沒有 LLM 特徵，則使用原始 DeBERTa 分類路徑
            logits = self.classifier(cls_output)


        # 5. 計算 Loss (標準 Hugging Face 邏輯)
        loss = None
        if labels is not None:
            loss_fct = nn.CrossEntropyLoss()
            loss = loss_fct(logits.view(-1, self.num_labels), labels.view(-1))

        return (loss, logits) if loss is not None else logits

num_labels = 2
MODEL_NAME = "microsoft/deberta-v3-base"

# 🌟 使用新定義的模型類別
model = DebertaWithLLMFeatures.from_pretrained(
    MODEL_NAME,
    num_labels=num_labels,
    trust_remote_code=True
)

# 🌟 將模型移動到 CUDA 裝置上
model.to(device)


def compute_metrics(eval_pred):
    # ... (指標計算保持不變)
    logits, labels = eval_pred
    preds = logits.argmax(axis=-1)
    acc = accuracy_score(labels, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(
        labels, preds, average='binary'
    )
    return {
        "accuracy": acc,
        "precision": precision,
        "recall": recall,
        "f1": f1,
    }

print("模型載入完成。")
```

```python
output_dir = "/content/deberta_llm_fusion_output"

training_args = TrainingArguments(
    output_dir=output_dir,
    learning_rate=2e-5,
    per_device_train_batch_size=32, # 批次大小 32
    per_device_eval_batch_size=64,
    num_train_epochs=3,
    weight_decay=0.01,
    eval_strategy="epoch",
    save_strategy="epoch",
    logging_dir="/content/logs",
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    greater_is_better=True,
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    tokenizer=tokenizer,
    data_collator=custom_data_collator, # 🌟 使用 CustomDataCollator
    compute_metrics=compute_metrics,
)

print("--- 開始訓練 DeBERTa + LLM Feature Fusion (GPU Optimized) ---")
trainer.train()

eval_result = trainer.evaluate()
print("\n最終驗證集結果 (使用最佳模型):", eval_result)
```

```python
# 對測試集做預測
# 由於 CustomDataCollator 已確保 llm_features 被傳遞，預測會自動使用融合後的模型
predictions = trainer.predict(test_dataset)
test_logits = predictions.predictions
test_preds = np.argmax(test_logits, axis=-1)

# 產生 submission 檔案
submission = sample_sub.copy()
submission["target"] = test_preds

output_path = "/content/submission_deberta_llm_fusion.csv"
submission.to_csv(output_path, index=False)

print(f"成功將提交檔案儲存至 {output_path}")
print("前 5 筆預測結果:")
print(submission.head())
```
